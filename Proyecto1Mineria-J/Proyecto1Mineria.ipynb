# Importacion de datos y librerias

Importamos las librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

Aqui se cargan los datos

```python
df = pd.read_csv("movies_cleaned.csv")
```

# Analisis Variables Cuantitativas

```python
variables_cuantitativas = [
    "budget", "revenue", "runtime", "voteAvg", "voteCount",
    "genresAmount", "productionCoAmount", "productionCountriesAmount",
    "actorsAmount", "castWomenAmount", "castMenAmount"
]
sns.set(style="whitegrid")

# Crear un grfico para cada variable cuantitativa
for var in variables_cuantitativas:
    if var in df.columns:
        plt.figure(figsize=(8, 6)) 
        sns.histplot(df[var].dropna(), kde=True, bins=30)  
        plt.title(f"Histograma de {var}", fontsize=14)  
        plt.xlabel(var, fontsize=12) 
        plt.ylabel('Frecuencia', fontsize=12) 
        plt.show()
```

Con estos histogramas podemos concluir que runtime y voteAVG son las unicas que cumplen y tienen una distribucion normal mientas que todas las demas tienen la forma de sesgada hacia la derecha, excepto GenresAmount ya que tiene una distribucion Multimodal.

# Analisis de Variables Cualitativas

```python
variables_cualitativas = [
    "genre1", "genre2", "genre3", "homePage", "productionCompany",
    "productionCompanyCountry", "productionCountry", "video", "director",
    "actors", "actorsCharacter", "originalTitle", "title", "originalLanguage"
]

# Mostrar tablas
for var in variables_cualitativas:
    if var in df.columns:
        frecuencias = df[var].value_counts()
        tabla_frecuencias = frecuencias.to_frame(name="Frecuencia")
        tabla_estilizada = tabla_frecuencias.style.set_properties(**{'text-align': 'left'})
        display(tabla_frecuencias)  
```

# Cules son las 10 pelculas que contaron con ms presupuesto?

```python

df["budget"] = pd.to_numeric(df["budget"], errors="coerce")

peliculas_ms_presupuesto = df.nlargest(10, "budget")

result = peliculas_ms_presupuesto [["id", "title", "budget"]]

from IPython.display import display

display(result.style.set_properties(**{'text-align': 'left'}).set_caption("Top 10 Pelculas con Mayor Presupuesto"))
```

# Cul es la pelcula que ms votos tuvo?

```python
df["voteCount"] = pd.to_numeric(df["voteCount"], errors="coerce")

pelicula_ms_votos = df.nlargest(1, "voteCount")

result = pelicula_ms_votos[["id", "title", "voteCount"]]

display(result.style.set_properties(**{'text-align': 'left'}).set_caption("Pelcula con Ms Votos"))
```

# Cuntas pelculas se hicieron en cada ao? En qu ao se hicieron ms pelculas?

```python
df["releaseDate"] = pd.to_datetime(df["releaseDate"], errors="coerce")
df["ao"] = df["releaseDate"].dt.year

peliculas_por_ao = df["ao"].value_counts().sort_index()
ao_mas_peliculas = peliculas_por_ao.idxmax()
cantidad_mas_peliculas = peliculas_por_ao.max()
print(f"El ao con ms pelculas fue {ao_mas_peliculas} con {cantidad_mas_peliculas} pelculas.")

# Grfico de barras
plt.figure(figsize=(12,6))
peliculas_por_ao.plot(kind="bar", color="skyblue")
plt.xlabel("Ao")
plt.ylabel("Cantidad de Pelculas")
plt.title("Cantidad de Pelculas por Ao")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

# Las pelculas de qu genero principal obtuvieron mayores ganancias?

```python
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")

ganancias_por_genero = df.groupby("genre1")["revenue"].sum().sort_values(ascending=False)
tabla_ganancias = ganancias_por_genero.reset_index()
tabla_ganancias.columns = ["Gnero", "Ganancias Totales"]
display(tabla_ganancias.style.set_properties(**{"text-align": "left"}).set_caption("Ganancias Totales por Gnero Principal"))
```

# Es posible que la cantidad de hombres y mujeres en el reparto influya en la popularidad y los ingresos de las pelculas?


```python
peliculas_df_limpio = df.dropna(subset=['castWomenAmount', 'castMenAmount', 'popularity', 'revenue'])

peliculas_df_limpio['reparto_total'] = peliculas_df_limpio['castWomenAmount'] + peliculas_df_limpio['castMenAmount']
peliculas_df_limpio['proporcion_mujeres'] = peliculas_df_limpio['castWomenAmount'] / peliculas_df_limpio['reparto_total']
peliculas_df_limpio['proporcion_hombres'] = peliculas_df_limpio['castMenAmount'] / peliculas_df_limpio['reparto_total']

correlacion_mujeres_popularidad = peliculas_df_limpio[['proporcion_mujeres', 'popularity']].corr().iloc[0, 1]
correlacion_hombres_popularidad = peliculas_df_limpio[['proporcion_hombres', 'popularity']].corr().iloc[0, 1]
correlacion_mujeres_ingresos = peliculas_df_limpio[['proporcion_mujeres', 'revenue']].corr().iloc[0, 1]
correlacion_hombres_ingresos = peliculas_df_limpio[['proporcion_hombres', 'revenue']].corr().iloc[0, 1]

plt.figure(figsize=(12, 20))

# Grfico 1: Proporcin de mujeres vs. popularidad
plt.subplot(2, 2, 1)
sns.scatterplot(data=peliculas_df_limpio, x='proporcion_mujeres', y='popularity', alpha=0.5)
plt.text(0.02, max(peliculas_df_limpio['popularity']), f'Correlacin: {correlacion_mujeres_popularidad:.2f}',
         fontsize=12)
plt.title('Proporcin de Mujeres vs. Popularidad')
plt.xlabel('Proporcin de Mujeres')
plt.ylabel('Popularidad')

# Grfico 2: Proporcin de hombres vs. popularidad
plt.subplot(2, 2, 2)
sns.scatterplot(data=peliculas_df_limpio, x='proporcion_hombres', y='popularity', alpha=0.5)
plt.text(0.02, max(peliculas_df_limpio['popularity']), f'Correlacin: {correlacion_hombres_popularidad:.2f}',
         fontsize=12)
plt.title('Proporcin de Hombres vs. Popularidad')
plt.xlabel('Proporcin de Hombres')
plt.ylabel('Popularidad')

# Grfico 3: Proporcin de mujeres vs. ingresos
plt.subplot(2, 2, 3)
sns.scatterplot(data=peliculas_df_limpio, x='proporcion_mujeres', y='revenue', alpha=0.5)
plt.text(0.02, max(peliculas_df_limpio['revenue']), f'Correlacin: {correlacion_mujeres_ingresos:.2f}',
         fontsize=12)
plt.title('Proporcin de Mujeres vs. Ingresos')
plt.xlabel('Proporcin de Mujeres')
plt.ylabel('Ingresos')

# Grfico 4: Proporcin de hombres vs. ingresos
plt.subplot(2, 2, 4)
sns.scatterplot(data=peliculas_df_limpio, x='proporcion_hombres', y='revenue', alpha=0.5)
plt.text(0.02, max(peliculas_df_limpio['revenue']), f'Correlacin: {correlacion_hombres_ingresos:.2f}',
         fontsize=12)
plt.title('Proporcin de Hombres vs. Ingresos')
plt.xlabel('Proporcin de Hombres')
plt.ylabel('Ingresos')

plt.tight_layout()
plt.show()
```

Despues de los analisis podemos concluir que la correlacion entre la popularidad de una pelicula vs cantidad hombres o mujeres esten no estan relacionados y afecte significativamente, y en cuanto la a los ingresos es igual la proporcion de genero masculino y femenino no afecta a los ingresos de una produccin. 

# Cmo se correlacionan los presupuestos con los ingresos? Los altos presupuestos significan altos ingresos?

```python

df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")

#histograma de presupuestos e ingresos
plt.figure(figsize=(12,6))
sns.histplot(df[["budget", "revenue"]], bins=50, kde=True, alpha=0.6)
plt.xlabel("Cantidad en dlares")
plt.ylabel("Frecuencia")
plt.title("Distribucin de Presupuestos e Ingresos")
plt.legend(["Presupuesto", "Ingresos"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

#diagrama de dispersin entre presupuesto e ingresos
correlation = df["budget"].corr(df["revenue"])
plt.figure(figsize=(12,6))
sns.scatterplot(x=df["budget"], y=df["revenue"], alpha=0.5)
plt.xlabel("Presupuesto")
plt.ylabel("Ingresos")
plt.title("Relacin entre Presupuesto e Ingresos")
plt.text(0.05, 0.95, f'Correlacin: {correlation:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

```

Segun los resultados del histograma que se puede ver un sesgo hacia la derecha indica la mayoria de proyectos estan en el rango bajo- medio en terminos e ingresos y que son pocos los proyectos los cuales tienen un gran presupuesto y por lo tanto un gran ingreso.
El diagrama de dispersion mostro una correlacion de 0.76 lo cual indica que hay una fuerte relacion entre los presupuestos e ingresos e indica que los proyectos con mayores presupuestos generan significativamente mas ingresos.

# En qu meses se han visto los lanzamientos con mejores ingresos? cuantas pelculas, en promedio, se han lanzado por mes?

```python
df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")
df["releaseDate"] = pd.to_datetime(df["releaseDate"], errors="coerce")
df["mes"] = df["releaseDate"].dt.month
ingresos_por_mes = df.groupby("mes")["revenue"].mean().sort_values(ascending=False)
peliculas_por_mes = df["mes"].value_counts().sort_index().mean()

tabla_ingresos = ingresos_por_mes.reset_index()
tabla_ingresos.columns = ["Mes", "Ingresos Promedio"]
display(tabla_ingresos.style.set_properties(**{"text-align": "left"}).set_caption("Ingresos Promedio por Mes"))

print(f"En promedio, se han lanzado {peliculas_por_mes:.2f} pelculas por mes.")
```

# La popularidad del elenco est directamente correlacionada con el xito de taquilla?

```python

popularity_columns = ["actorPopularity1", "actorPopularity2", "actorPopularity3", "actorPopularity4", "actorPopularity5"]
df[popularity_columns] = df[popularity_columns].apply(pd.to_numeric, errors="coerce")
df["castPopularity"] = df[popularity_columns].mean(axis=1)
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")

#Diagrama de dispersin entre popularidad del elenco e ingresos
plt.figure(figsize=(12,6))
sns.scatterplot(x=df["castPopularity"], y=df["revenue"], alpha=0.5)
plt.xlabel("Popularidad del Elenco")
plt.ylabel("Ingresos de Taquilla")
plt.title("Relacin entre Popularidad del Elenco e Ingresos de Taquilla")
correlacion = df["castPopularity"].corr(df["revenue"])
plt.text(0.05, 0.95, f'Correlacin: {correlacion:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

La correlacion de la tabla anterior demuestra que la correaccion es debil siendo de 0.37 por lo tanto la popularidad de los actores no influye en los ingresos en taquilla.
